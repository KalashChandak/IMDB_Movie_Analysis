```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```
df.drop(columns=['Poster_Link','Overview'],inplace=True)
```

We load the IMDb Top 1000 movies dataset using pandas.

```
df.head()
```

```
df.info()
```

```
# Counting missing values in each column
missing_counts = pd.isnull(df).sum()
print(missing_counts)
```

CLEAN THE DATASET

```
#Rename columns for convenience
df.columns = df.columns.str.strip().str.replace(' ','_').str.replace('.','_')

#Remove 'min' from Runtime and replace with '' and convert to integer
df['Runtime']=df['Runtime'].str.replace('min','').astype(float)

#Convert gross to numeric(remove commas and convert)
df['Gross']=df['Gross'].replace('[^\d.]','',regex=True).replace('',np.nan).astype(float)

#Convert release year to numeric
df['Released_Year']=pd.to_numeric(df['Released_Year'],errors='coerce')

df.head()
```

HANDLING MISSING VALUES

```
df.isnull().sum()
```

```
#Dropping the row where released_year is missing value
df.dropna(subset=['Released_Year'], inplace=True)
```

```
#fill null values in meta_score using mean and in gross using median
df = df.copy()  # safe step if df is a filtered version

df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].mean())
df['Gross'] = df['Gross'].fillna(df['Gross'].median())
```

```
#Replacing null values in certificate with 'not rated'
df['Certificate'] = df['Certificate'].fillna('Not Rated')
```

```
df.isnull().sum()
```

##FEATURE ENGINEERING

```
df['Decade'] = (df['Released_Year'] // 10 * 10).astype('Int64')
df['Main_Genre'] = df['Genre'].str.split(',').str[0]
```

```
df.head()
```

```
df.isnull().sum()
```

# DATA VISUALIZATION(EDA)

```
#Genre vs Ratings vs Votes

bubble_df = df.groupby('Main_Genre').agg({'IMDB_Rating':'mean','No_of_Votes':'mean','Gross':'mean'}).reset_index()
fig = px.scatter(bubble_df, x='IMDB_Rating', y='Gross', size='No_of_Votes', 
                 color='Main_Genre', title='Genre Bubble Chart (Rating vs Gross vs Popularity)')
fig.show()
```

INTERPRETATION:
1.Action & Animation Lead Across the Board
2.Family Genre: High Gross, Lower Rating/Popularity
3.Drama & Comedy: Critically Decent, Lower Gross
4.High Ratings Don't Always Equal High Gross

```
#Certificate vs IMDB Rating

plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='Certificate',y='IMDB_Rating',palette='Set3')
plt.title('IMDb Rating Distribution by Certificate')
plt.show()
```

### üéüÔ∏è Insight: Certificate vs IMDb Rating

Movies rated U/A and R show a wide spread in IMDb ratings. While U-rated films maintain consistent family-friendly quality, R-rated films exhibit both critically acclaimed and average entries, reflecting genre diversity.


```
#Genre Distribution

from collections import Counter

# Create a flat list of all genres
genre_list = df['Genre'].dropna().apply(lambda x: [g.strip() for g in x.split(',')])
all_genres = [g for sublist in genre_list for g in sublist]

# Count occurrences
genre_counts = Counter(all_genres)
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index').reset_index()
genre_df.columns = ['Genre', 'Count']

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=genre_df.sort_values(by='Count', ascending=False), x='Genre', y='Count', palette='pastel')
plt.title('Total Occurrences of Each Genre (Split)')
plt.xticks(rotation=45)
plt.show()
```

### üé≠ Insight: Genre Distribution

Drama, Comedy, and Action dominate the top 1000 movie genres. This reflects audiences‚Äô strong engagement with emotionally driven narratives, humor, and thrilling storylines.


```
# Correlation for numeric features

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Enhanced Correlation Matrix')
plt.show()
```

### üìà Insight: Correlation Between Numerical Features

IMDb Rating, Metascore, and Number of Votes show moderate correlations, indicating that critical reception and popularity often align. Gross earnings, however, are less correlated with IMDb ratings, suggesting that box office success doesn't guarantee high critical acclaim.


```
sns.pairplot(df[['IMDB_Rating', 'Meta_score', 'Gross', 'Runtime']], corner=True)
plt.show()
```

### üîç Insight: Pairplot ‚Äî IMDb Rating, Metascore, Gross, and Runtime

The pairplot reveals key relationships among numerical features:
- A **moderate positive correlation** exists between IMDb rating and Metascore, showing that audience and critic opinions often align.
- **Gross earnings** show no strong direct correlation with IMDb or Metascore, indicating that financial success doesn't always reflect critical or audience reception.
- **Runtime** slightly trends upward with higher IMDb ratings, suggesting that longer, more in-depth films may receive better reviews ‚Äî though the effect is mild.

These insights help identify variables that influence perceived and commercial success.


```
#Gross Revenue by Decades

sns.set_style('darkgrid')
df['Decade'] = (df['Released_Year'] // 10) * 10
df.groupby('Decade')['Gross'].mean().plot(kind='bar', figsize=(10,5), color='goldenrod')
plt.title('Average Gross Revenue by Decade')
plt.ylabel('Average Gross (in USD)')
plt.xticks(rotation=45)
plt.show()
```

### üí∞ Insight: Average Gross by Decade

The 2010s saw the highest average gross, showcasing the financial power of modern-day franchises, international markets, and inflation-adjusted blockbuster strategies.

```
#Top Actors

from collections import Counter

actor_list = df['Star1'].tolist() + df['Star2'].tolist() + df['Star3'].tolist() + df['Star4'].tolist()
actor_counts = pd.Series(Counter(actor_list)).sort_values(ascending=False).head(10)

actor_counts.plot(kind='bar', figsize=(10,5), color='salmon')
plt.title('Top 10 Most Frequent Actors in IMDb Top 1000')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

### ‚≠ê Insight: Most Frequent Actors

Leonardo DiCaprio, Tom Hanks, and Robert De Niro frequently appear, indicating consistent success and popularity in critically acclaimed films over decades.


```
#GENRE PROPORTION PER DECADE

# Create decade column
df['Decade'] = (df['Released_Year'] // 10) * 10

# Get count of genres per decade
genre_decade = df.groupby(['Decade', 'Main_Genre']).size().unstack().fillna(0)

# Normalize to get proportions
genre_decade_percent = genre_decade.div(genre_decade.sum(axis=1), axis=0)

# Plot
genre_decade_percent.plot(kind='bar', stacked=True, colormap='tab20', figsize=(14,7))
plt.title("Genre Distribution by Decade (Proportion)")
plt.xlabel("Decade")
plt.ylabel("Proportion")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### üìΩÔ∏è Insight: Genre Proportion per Decade

The proportion of movie genres has shifted notably across decades. Drama and Romance were dominant in earlier eras, while Action, Adventure, and Sci-Fi gained significant traction post-1990s ‚Äî aligning with advancements in visual effects and audience appetite for spectacle. Comedy has remained relatively steady, reflecting its evergreen appeal across generations.


```
#TOP DIRECTORS BY AVERAGE RATING(>= 3 MOVIES)

dir_rating = df.groupby('Director')['IMDB_Rating'].agg(['mean','count'])
dir_rating = dir_rating[dir_rating['count'] >= 3].sort_values('mean', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=dir_rating['mean'], y=dir_rating.index, palette='viridis')
plt.title("Top 10 Directors by Avg IMDb Rating (‚â• 3 Movies)")
plt.xlabel("Avg Rating")
plt.ylabel("Director")
plt.show()
```

### üé¨ Insight: Top Directors by Average IMDb Rating

Directors like Christopher Nolan, Quentin Tarantino, and Martin Scorsese consistently maintain high average IMDb ratings, reflecting both strong directorial style and audience appeal. Their films often blend unique storytelling with technical excellence, contributing to their critical acclaim.


```
#IMDB vs METASCORE_GAP (MOST DISAGREED MOVIES)

df['Rating_Diff'] = df['IMDB_Rating'] - (df['Meta_score'] / 10)

# Most overrated by public (IMDb >> Metascore)
overrated = df.sort_values('Rating_Diff', ascending=False).head(10)[['Series_Title', 'IMDB_Rating', 'Meta_score', 'Rating_Diff']]

# Most underrated by public (Metascore >> IMDb)
underrated = df.sort_values('Rating_Diff', ascending=True).head(10)[['Series_Title', 'IMDB_Rating', 'Meta_score', 'Rating_Diff']]

display(overrated)
display(underrated)
```

### üßÆ Insight: IMDb vs Metascore Gap

While IMDb ratings reflect general audience sentiment, Metascores capture critic reviews. A noticeable gap between the two often highlights divisive films ‚Äî those loved by audiences but not critics (e.g., action-heavy or nostalgic films), and vice versa for slower, arthouse cinema. Most films maintain moderate alignment, but outliers signal differences in perception and expectations.


```
#Average runtime per decade

avg_runtime_per_decade = df.groupby('Decade')['Runtime'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_runtime_per_decade, x='Decade', y='Runtime', marker='o')
plt.title('Average Movie Runtime per Decade')
plt.xlabel('Decade')
plt.ylabel('Average Runtime (minutes)')
plt.grid(True)
plt.show()
```

## Insight on Average Movie Runtime per Decade
The graph illustrates a significant increase in average movie runtimes from 1920 to 1960 (from ~86 to ~128 minutes). After a slight dip and stabilization in the 1970s and 1980s, runtimes have largely remained high, with another peak around the late 2000s, indicating a general trend towards longer films over the past century.
